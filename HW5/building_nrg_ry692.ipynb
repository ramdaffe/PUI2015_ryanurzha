```python
nrg['BBL'] = nrg['NYC Borough, Block, and Lot (BBL)']
bblnrgdata = pd.merge(nrg, bsize, on='BBL')
bblnrgdata
```

```python
#key column:  u'UnitsTotal', Reported Property Floor Area (Building(s)) (ft²)',Site EUI(kBtu/ft2)
print bblnrgdata['Site EUI(kBtu/ft2)'].head()
print bblnrgdata['Reported Property Floor Area (Building(s)) (ft²)'].head()

bblnrgdata["TotalEnergy"] = bblnrgdata['Site EUI(kBtu/ft2)'].astype(float) * bblnrgdata['Reported Property Floor Area (Building(s)) (ft²)'].astype(float)
bblnrgdata["TotalUnits"] = bblnrgdata["UnitsTotal"][(bblnrgdata["UnitsTotal"] > 0)]
```

```python
fig = pl.figure(7,figsize=(10,5))
ax = fig.add_subplot(111)
ax.plot(bblnrgdata["TotalUnits"], bblnrgdata["TotalEnergy"], 'o', color='SteelBlue')
ax.set_xlabel("Number of units",color='gray', fontsize=16)
ax.set_ylabel("Total Energy",color='gray', fontsize=16)
ax.set_yscale('log')
ax.set_xlim([0, 1000])
#ax.set_xscale('log')
```

```python
bblnrg_cut = bblnrgdata[(bblnrgdata.TotalEnergy > 1000) & (bblnrgdata.TotalEnergy < 1e10) & (bblnrgdata.TotalUnits>=10) & (bblnrgdata.TotalUnits<1000)]
bblnrg_cut.plot(kind='scatter',x='TotalEnergy',y='TotalUnits', marker='o',  figsize=(14, 14), loglog=True)
```

```python
import statsmodels.api as sm
X = sm.add_constant(np.log10(bblnrg_cut.TotalUnits))
model = sm.OLS(np.log10(bblnrg_cut.TotalEnergy), X, missing='drop') # ignores entires where x or y is NaN
fit = model.fit()
print fit.summary()
pl.figure(figsize=(14, 14))
pl.scatter(np.log10(bblnrg_cut.TotalUnits), np.log10(bblnrg_cut.TotalEnergy))
plot(np.log10(bblnrg_cut.TotalUnits), fit.predict(), 'k')
pl.ylabel('Total Energy Consumption')
pl.xlabel('Total Units')
```
